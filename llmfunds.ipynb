```python
import json
import pandas as pd

fixed_data = []


with open("mutual_funds_data.json", "r") as f:
    for line in f:
        line = line.strip()
        if line:
            try:
                fixed_data.append(json.loads(line))
            except json.JSONDecodeError as e:
                print("Error loading line:", e)

# Convert to DataFrame
df = pd.DataFrame(fixed_data)

# Save to CSV
df.to_csv("fixed_mutual_funds_data.csv", index=False)

print("Done! fixed_mutual_funds_data.csv file is ready.")
```

```python
import pandas as pd

# Load your cleaned fund dataset
df = pd.read_csv("fixed_mutual_funds_data.csv")

# Fill NaNs just in case
df = df.fillna("")

# Create a combined text field for each fund
def combine_metadata(row):
    return f"{row['schemeName']} {row['amcName']} {row['category']} {row['subCategory']} {row['riskOMeter']} {row['objective']}"

df["combined_text"] = df.apply(combine_metadata, axis=1)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(df["combined_text"].tolist(), show_progress_bar=True)
```

```python
print(embeddings.shape)
```

```python
 import logging
logging.getLogger("sentence_transformers").setLevel(logging.ERROR)
```

```python
import faiss

# Create FAISS index
dimension = embeddings[0].shape[0]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

print("Vector index created with", index.ntotal, "funds")
```

```python
import faiss

# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

# Add embeddings to index
index.add(np.array(embeddings))

# Save index to disk (optional)
faiss.write_index(index, "fund_index.faiss")
```

```python
def search_fund(query, top_k=10):
    query_embedding = model.encode([query])
    D, I = index.search(np.array(query_embedding), top_k)
    
    print("\n🔍 Top Matches for:", query)
    for i in I[0]:
        print(f"\n🎯 {df.iloc[i]['schemeName']}")
        print(f"AMC: {df.iloc[i]['amcName']}")
        print(f"Category: {df.iloc[i]['category']} > {df.iloc[i]['subCategory']}")
        print(f"Risk: {df.iloc[i]['riskOMeter']}")
        print(f"Why: Matched on -> {df.iloc[i]['combined_text'][:150]}...")
```

```python
search_fund("hdfc")
```
